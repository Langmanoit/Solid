```python
import numpy as np
import matplotlib.pyplot as plt

# Задаем набор точек
x = np.array(x2[:len(x2)//1])
y = np.array(y2[:len(x2)//1])

# Задаем порядок многочлена
n = 100

# Нормализуем данные
x_mean = np.mean(x)
x_std = np.std(x)
y_mean = np.mean(y)
y_std = np.std(y)

x_normalized = (x - x_mean) / x_std
y_normalized = (y - y_mean) / y_std

# Фитируем многочлен заданного порядка
p = np.polyfit(x_normalized, y_normalized, n)
poly = np.poly1d(p)

# Создаем данные для графика
x_fit = np.linspace(min(x), max(x), 10000)
y_fit = poly((x_fit - x_mean) / x_std) * y_std + y_mean

# Строим график многочлена
plt.figure(figsize=(17, 8))
plt.scatter(x, y, label='Исходные точки')
plt.xlim((max(x)-min(x))/100+ min(x),max(x)-(max(x)-min(x))/100)
plt.ylim(0.05, 0.11)
plt.plot(x_fit, y_fit, color='red', label='Приближающий многочлен')
plt.xlabel('X')
plt.ylabel('Y')
plt.legend()
plt.grid()
plt.show()

# Получаем производную многочлена
derivative = np.polyder(p)

# Генерируем значения x для построения графика производной
x_values = np.linspace(min(x), max(x), 10000)
y_derivative = np.polyval(derivative, (x_values - x_mean) / x_std) * y_std

# Строим график производной
plt.figure(figsize=(17, 8))
plt.plot(x_values, y_derivative, label='Производная многочлена')
plt.xlabel('X')
plt.ylabel('Y')
plt.legend()
plt.grid()
plt.ylim(-0.5, 0.5)
plt.show()
```
